```python
account_balance = 0
pin = 1234

def view_balance():
  print('Your balance is',account_balance)
  global user_input
  user_input = int(input('\nPlease select an option below:\n1. View Account Balance\n2. Deposit Money\n3. Withdraw Money\n4. Exit\n'))
  menu(user_input)

def deposit():
  deposit_amount = int(input('Enter amount to be deposited:'))
  global account_balance
  account_balance += deposit_amount
  print('Your new balance is:',account_balance)
  global user_input
  user_input = int(input('\nPlease select an option below:\n1. View Account Balance\n2. Deposit Money\n3. Withdraw Money\n4. Exit\n'))
  menu(user_input)

def withdraw():
  options = input('Enter amount to be withdrawn:\n10\n20\n50\n100\n200\n500\nOther amount (Type \'Other\')\n')
  while options not in ('10','20','50','100','200','500') and options != 'Other':
    options = input('\nSelect a valid option::\n10\n20\n50\n100\n200\n500\nOther amount (Type \'Other\')\n')
  if options == 'Other':
    withdraw_amount = int(input('Enter the amount:'))
    while withdraw_amount %10 != 0:
      withdraw_amount = int(input('\nPlease enter a multiple of 10:'))
  elif options in ('10','20','50','100','200','500'):
    withdraw_amount = int(options)
  global account_balance
  while account_balance - withdraw_amount < 0:
    print('\nPlease select a lower amount')
    withdraw()
  else:
    account_balance -= withdraw_amount
    print('Your new balance is:',account_balance)
  global user_input
  user_input = int(input('\nPlease select an option below:\n1. View Account Balance\n2. Deposit Money\n3. Withdraw Money\n4. Exit\n'))
  menu(user_input)

def exit():
  print('Successfully exited')

def menu(x):
  if x == 1:
    view_balance()
  elif x == 2:
    deposit()
  elif x == 3:
    withdraw()
  elif x == 4:
    exit()
  else:
    user_input = int(input('Invalid option, please enter again'))
    menu(user_input)

pin_input = int(input('Enter your PIN:'))
while pin_input != pin:
  pin_input = int(input('Incorrect, please enter your PIN again:'))
else:
  user_input = int(input('\nPlease select an option below:\n1. View Account Balance\n2. Deposit Money\n3. Withdraw Money\n4. Exit\n'))
  menu(user_input)
```
